```python
import os as os
import csv
import string
import pandas as pd
import numpy as np
```

```python
lease_csv = os.path.join("datagovlserexus.csv")
lease_df = pd.read_csv(lease_csv)
```

```python
lease_df.head()
```

```python
lease_df["CleanZipCode"] = np.zeros(len(lease_df))
```

```python
for i in range(len(lease_df)):
    string = lease_df["Lessor Zip Code"][i]
    lease_df["CleanZipCode"][i] = string[0:5]
```
